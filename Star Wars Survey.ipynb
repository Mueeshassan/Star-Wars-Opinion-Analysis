## Calculating the percentage of males and females that have seen the Star Wars Movies

```python
star_wars['seen_all'] = star_wars[star_wars.columns[3:9]].all(1, skipna=True)
star_wars['seen_all'] = star_wars['seen_all'].fillna(False)

star_wars['seen_one'] = star_wars[star_wars.columns[3:9]].any(1, skipna=True)
star_wars['seen_one'] = star_wars['seen_one'].fillna(False)

def percentage_val(x):
    return round(x.sum() / x.count() * 100, 1)
seen_all = star_wars.pivot_table(index ='Gender', values=['seen_all'], aggfunc=percentage_val)
seen_one = star_wars.pivot_table(index ='Gender', values=['seen_one'], aggfunc=percentage_val)
```

## Visualizing the results:
#### The first plot shows the percentage of males and females from the survey that have seen all the star wars series
#### The second plot shows the percentage of males and females from the survey that have seen at least one of the series

```python
seen_all.plot(kind='bar', color= 'blue', alpha=0.7)
ax = plt.gca()
plt.title('Seen: All Movies', weight='bold')
plt.xticks(rotation=0)
ax.set_ylabel('Seen (%)', fontsize=12, alpha=0.7)
ax.spines['left'].set_linewidth(0.1)
ax.spines['bottom'].set_linewidth(0.1)
ax.spines['top'].set_linewidth(0.1)
ax.spines['right'].set_linewidth(0.1)
plt.tick_params(axis='both', which='both', length=0)
plt.legend([])
plt.show()

seen_one.plot(kind='bar', color= 'blue', alpha=0.7)
ax = plt.gca()
plt.title('Seen: At Least One Movie', weight='bold')
plt.xticks(rotation=0)
ax.set_ylabel('Seen (%)', fontsize=12, alpha=0.8)
ax.spines['left'].set_linewidth(0.1)
ax.spines['bottom'].set_linewidth(0.1)
ax.spines['top'].set_linewidth(0.1)
ax.spines['right'].set_linewidth(0.1)
plt.tick_params(axis='both', which='both', length=0)
plt.legend([])
plt.show()
```

## Analyzing the opinion on 'who shot first?' 

"Who shot first?" is a phrase that refers to a famous controversy in the Star Wars community, particularly concerning a scene in "Star Wars Episode IV: A New Hope."

In the original 1977 release of the film, there is a scene in the Mos Eisley Cantina where Han Solo (played by Harrison Ford) is confronted by the bounty hunter Greedo. In this version, Han shoots Greedo first without any return fire from Greedo. Many fans interpreted this action as Han acting in self-defense and being a more morally ambiguous character.

However, when George Lucas, the creator of Star Wars, made some changes to the movie in later releases (Special Editions), he altered the scene to show Greedo shooting at Han first, and then Han returning fire. This change was made to make Han Solo appear less morally ambiguous and to show that he was reacting to a threat.

The alteration sparked debates among fans, with some arguing that the change fundamentally altered Han Solo's character, while others felt it was a trivial issue.

The phrase "Who shot first?" has since become a shorthand for discussing changes and controversies in Star Wars and represents how small details in a film can hold significant meaning and impact for fans.

## Analyzing the opinion on 'who shot first?' based on age

```python
star_wars['shot_first'] = star_wars['Which character shot first?']
age_shot_first = star_wars.pivot_table(index= 'Age', values= 'Which character shot first?', columns= 'shot_first', aggfunc='count')
age_shot_first_han = age_shot_first['Han']
age_shot_first_greedo = age_shot_first['Greedo']
age_shot_first_none = age_shot_first["I don't understand this question"]
age_shot_first_han.plot(kind='bar', position=1, width=0.14, label='Han')
age_shot_first_greedo.plot(kind='bar',position=0, width=0.14, color='green', label='Greedo')
age_shot_first_none.plot(kind='bar',position=2, width=0.14, color='orange', label='do not understand')
ax = plt.gca()
plt.title('Opinion: Who Shot First?', weight='bold')
ax.spines['left'].set_linewidth(0.1)
ax.spines['bottom'].set_linewidth(0.1)
ax.spines['top'].set_linewidth(0.1)
ax.spines['right'].set_linewidth(0.1)
ax.set_ylabel('Total Number')
plt.xticks(fontsize= 12, alpha=0.7, rotation=0)
plt.yticks(alpha=0.7)
plt.tick_params(axis='both', which='both', length=0)
plt.legend()
plt.show()
```
